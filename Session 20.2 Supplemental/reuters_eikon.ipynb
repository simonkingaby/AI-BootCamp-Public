````python
import eikon as ek
import os
import dotenv
import pandas as pd

dotenv.load_dotenv()

ek.set_app_key(os.getenv('EK_API_KEY'))
````

````python
# Default top 10 English language news headlines
headlines_df = ek.get_news_headlines()
print(len(headlines_df), headlines_df.head())

# All news headlines for Lufthansa for the specific date and time range
headlines_df = ek.get_news_headlines('R:LHAG.DE', date_from='2019-03-06T09:00:00', date_to='2019-03-06T18:00:00')
print(len(headlines_df), headlines_df.head())

# First news headline for European Union and Politics
headlines_df = ek.get_news_headlines('EU AND POL',1)
print(len(headlines_df), headlines_df.head())

story_id = headlines_df.iat[0,2]
ek.get_news_story(story_id)
````
