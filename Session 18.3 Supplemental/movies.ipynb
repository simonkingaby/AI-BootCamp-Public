```python
import tmdbsimple as tmdb
import os
import pandas as pd

# import the env
from dotenv import load_dotenv
load_dotenv()
tmdb.API_KEY = os.getenv("TMDB_API_KEY")
```

```python
tmdb.REQUESTS_TIMEOUT = (2, 5)  # seconds, for connect and read specifically 
import requests
tmdb.REQUESTS_SESSION = requests.Session()
```

```python
def find_person(name):
    search = tmdb.Search()
    response = search.person(query=name)
    people = []
    for s in search.results:
        print(s['name'], s['id'])
        people.append({'name': s['name'], 'id': s['id']})
    return people

# find all the titles with a given person in them
def movies_by_person(id):
    person = tmdb.People(id)
    response = person.movie_credits()
    movies = []
    for credit in person.movie_credits()['cast']:
        print(credit['title'], credit['release_date'], credit['id'], credit['character'])
        movies.append({'title': credit['title'], 'release_date': credit['release_date'], 'id': credit['id'], 'character': credit['character']})
    return movies

movies = []
tom_hanks_movies = movies_by_person(find_person('Tom Hanks')[0]['id'])
print(tom_hanks_movies)
print(len(tom_hanks_movies))
movies.append({'name': 'Tom Hanks', 'movies': tom_hanks_movies, 'count': len(tom_hanks_movies)})
whoopi_goldberg_movies = movies_by_person(find_person('Whoopi Goldberg')[0]['id'])
print(whoopi_goldberg_movies)
print(len(whoopi_goldberg_movies))
movies.append({'name': 'Whoopi Goldberg', 'movies': whoopi_goldberg_movies, 'count': len(whoopi_goldberg_movies)})
johnny_depp_movies = movies_by_person(find_person('Johnny Depp')[0]['id'])
print(johnny_depp_movies)
print(len(johnny_depp_movies))
movies.append({'name': 'Johnny Depp', 'movies': johnny_depp_movies, 'count': len(johnny_depp_movies)})
kevin_bacon_movies = movies_by_person(find_person('Kevin Bacon')[0]['id'])
print(kevin_bacon_movies)
print(len(kevin_bacon_movies))
movies.append({'name': 'Kevin Bacon', 'movies': kevin_bacon_movies, 'count': len(kevin_bacon_movies)})
robin_williams_movies = movies_by_person(find_person('Robin Williams')[0]['id'])
print(robin_williams_movies)
print(len(robin_williams_movies))
movies.append({'name': 'Robin Williams', 'movies': robin_williams_movies, 'count': len(robin_williams_movies)})

```

```python
movies_df = pd.DataFrame(movies)
print(movies_df[['name', 'count']])
```
